## 1. Basic convolution [40 marks]

Implement basic convolution by translating the pseudo-code below to Python.

Fill in the `basic_convolution(image, kernel)` below, which takes as input a grayscale `image` (2D matrix) and a filtering `kernel` (2D matrix), and returns the convolved image result as a greyscale image with the same size and datatype as the input image.

This operation can be described by the following pseudocode:
```
for each image row in input image:
    for each pixel in image row:
        set accumulator to zero
        
        for each kernel row in kernel:
            for each element in kernel row:
                if element position <corresponding to> pixel position then
                    multiply element value <corresponding to> pixel value
                    add result to accumulator
                endif

        set output image pixel to accumulator
```
(`<corresponding to>` input image pixels are found relative to the kernel’s origin)

Source: https://en.wikipedia.org/wiki/Kernel_(image_processing)#Convolution

````python
def basic_convolution(image, kernel, verbose=False):
    'Computes the convolution of an image with a kernel.'
    # flip kernel
    kernel = np.flip(kernel)
    
    new_image = np.ones_like(image)
    ker_centre = [int((len(kernel)+1)/2 - 1), int((len(kernel[0])+1)/2 - 1)] # [row, column] of centre of Kernel

    # performs convolution on each pixel in image
    for imrow in range(len(image)): # for each row of the image
        for pixel in range(len(image[imrow])): # for each pixel in each row of the image
            accumulator = 0 # stores value of new pixel
            valid = True #tracks if current pixel can be kernelled
            for kerow in range(len(kernel)):
                if not valid: # leave loop if cant kernel
                    break
                for element in range(len(kernel[kerow])):
                    if len(image) > imrow - (ker_centre[0] - kerow) >= 0 and len(image[0]) > pixel - (ker_centre[1] - element) >= 0: # if kernel element maps onto image
                        accumulator += kernel[kerow, element] * image[imrow - (ker_centre[0] - kerow), pixel - (ker_centre[1] - element)] # add on element * pixel
                    else: # if kernel element does not correspond to a pixel, set new pixel as 0
                        valid = False
                        accumulator = 0
                        break
            new_image[imrow, pixel] = accumulator # set new pixel
    return new_image
````

## 2.	Border handling [20 marks]

Improve your implementation from the previous exercise:
1. Centre the filtered image, so that the content of the input and filtered images is not shifted between them; and
2. Fill in the border regions by extending/replicating the edge pixels of the image (‘clamp-to-edge’).

````python
def extended_convolution(image, kernel, verbose=False):
    image = basic_convolution(image, kernel, verbose = False)
    kernel_centre = [int((len(kernel)+1)/2 - 1), int((len(kernel[0])+1)/2 - 1)] #to store sizes of black zone
    for row in range(0, kernel_centre[0]): # sets top row to pixels below
        image[row, kernel_centre[1]: -kernel_centre[1]] = image[kernel_centre[0],kernel_centre[1]: -kernel_centre[1]]
        for pixel in range(0, kernel_centre[1]): # sets left corner to closest pixel
            image[row,pixel] = image[kernel_centre[0], kernel_centre[1]]
        for pixel in range(-kernel_centre[1], 0): # sets right corner to closest pixel
            image[row, pixel] = image[kernel_centre[0], -kernel_centre[1] - 1]
    for row in range(-kernel_centre[0], 0): # sets bottom row to pixels above
        image[row, kernel_centre[1]:-kernel_centre[1]] = image[-kernel_centre[0] - 1, kernel_centre[1]: -kernel_centre[1]]
        for pixel in range(0, kernel_centre[1]): # sets left corner to closest pixel
            image[row,pixel] = image[-kernel_centre[0] - 1, kernel_centre[1]]
        for pixel in range(-kernel_centre[1], 0): # sets right corner to closest pixel
            image[row, pixel] = image[-kernel_centre[0] - 1, -kernel_centre[1] - 1]
    for column in range(0, kernel_centre[1]): # sets left column to pixels to the right
        image[kernel_centre[0]: -kernel_centre[0], column] = image[kernel_centre[0]: -kernel_centre[0],kernel_centre[1]]
    for column in range(-kernel_centre[1], 0): # sets right column to pixels to the left
        image[kernel_centre[0]: -kernel_centre[0], column] = image[kernel_centre[0]: -kernel_centre[0], -kernel_centre[1] - 1]
    return image
````

### Test your result

A perfect result will match SciPy’s function `convolve` (with `mode='nearest'`) exactly, i.e. with a sum of squared differences (SSD) of 0 (zero):

````python
kernel = np.random.rand(5, 5)
kernel = kernel / kernel.sum()  # normalise kernel

filtered = extended_convolution(image_crop, kernel)
reference = scipy.ndimage.convolve(image_crop, kernel, mode='nearest')
ssd = np.sum(np.power(filtered - reference, 2))

plt.subplot(1, 3, 1); plt.title('Extended convolution');        plt.imshow(filtered,  cmap='gray', vmin=0,  vmax=1)
plt.subplot(1, 3, 2); plt.title('Reference result');            plt.imshow(reference, cmap='gray', vmin=0,  vmax=1)
plt.subplot(1, 3, 3); plt.title(f'Difference (SSD={ssd:.3f})'); plt.imshow(filtered - reference,   vmin=-1, vmax=1);
````

b) Compute the kernel for a 5×5 Gaussian blur with a standard deviation of 1 pixel.
Make sure that the kernel is normalised.

````python
## TASK: 5x5 Gaussian kernel with standard deviation of 1 pixel
gauss = np.zeros([5, 5])
gauss[2, 2] = 1 # make centre 1
std = 1.0 # standard deviation

#because gaussians have the centre as the origin, calculate offets against centre vs corner
offsetrow = -(len(gauss) - 1)/2 
offsetcolumn = -(len(gauss[0]) -1)/2 

# calculates gauss of each pixel (see slides for formula)
for row in range(len(gauss)):
    for element in range(len(gauss[row])):
        gauss[row,element] = (1/(2*np.pi*std**2))*(np.e)**(-((row+offsetrow)**2 + (element + offsetcolumn)**2)/(2*std**2))
        
#normalise gaussian
gauss = (1/gauss.sum()) * gauss

print("Gaussian 5x5 kernel:"); print(gauss)
print(f"\nSum of elements: {gauss.sum()}")
gauss_result = scipy.ndimage.convolve(image_crop, gauss, mode='nearest')
plt.title('Gaussian filtering'); plt.imshow(gauss_result, cmap='gray', vmin=0, vmax=1);
````

## 4.	Exploiting the convolution theorem [30 marks]

Apply the convolution theorem to speed up the convolution.

You can use NumPy’s functions for 2D Fast Fourier Transform (FFT), `np.fft.fft2` and its inverse `np.fft.ifft2`.

````python
def fft_convolution(image, kernel):
    # store original dimensions
    rows = len(image)
    columns = len(image[0])
    
    # match parity(odd/even) of image and kernel dimensions
    if rows % 2 != len(kernel) % 2:
        image = np.pad(image, ((0,1),(0,0))) # pad bottom row with blanks
    if columns % 2 != len(kernel[0]) %2:
        image = np.pad(image, ((0,0),(0,1))) #pad left column with blanks
        
    # pad kernel to match image size
    kernel = np.pad(kernel, (int(1/2 * (len(image) - len(kernel))), int(1/2*(len(image[0]) - len(kernel[0])))))
    
    # centre kernel in fourier domain
    kernel = np.fft.ifftshift(kernel)
    
    # fft on image and kernel
    imageft = np.fft.fft2(image)
    kernelft = np.fft.fft2(kernel)
    
    # fourier convolution
    con_imageft = (imageft * kernelft)
    
    # revert back to original form, and take only real values
    con_image = np.fft.ifft2(con_imageft).real
    
    # crop image to remove padding
    return con_image[0:rows, 0:columns]
````
